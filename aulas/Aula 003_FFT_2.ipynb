```python
import numpy as np, scipy as sp, matplotlib.pyplot as plt

T = 0.2 # periodo de amostragem
t0 = 0 # tempo inicial
tf = 100 # tempo final

# vetor tempo
t = np.arange(t0, tf, T)
# vetor de amplitudes no dominio do TEMPO
x = np.sin(0.44*np.pi*t) + 0.5*np.sin(np.pi*t) #senoide

# grafico no dominio do tempo
plt.figure(figsize=(8,2))
plt.plot(t, x);
plt.xlabel("tempo (s)")
plt.ylabel("$x(t)$");
```

```python
# implementando a FFT de x
from scipy.fft import fft

N = len(x) # numero amostras do vetor x
n_fft = 500 # numero de pontos da FFT
# calcula FFT
Xf = fft(x, n=n_fft)
Xf = Xf/N #correcao na amplitude na freq
Xf = Xf[0:n_fft//2] # elimina espelhamento no dom freq
Xf[1:] = 2*Xf[1:] # corrige alteração na amplitude

# obtem modulo e fase da FFT
mag = np.abs(Xf)
phase = np.angle(Xf)

# # vetor de frequencias [rad/amostra]
# w0 = 2*np.pi/n_fft
# freqs = np.arange(0, 2*np.pi, w0)

# # vetor de frequencias [rad/s]
# w0 = 2*np.pi/(T*n_fft)
# freqs = np.arange(0, n_fft*w0, w0)

# vetor de frequencias [Hz]
f0 = 1/(T*n_fft)
freqs = np.arange(0, n_fft*f0, f0)

freqs = freqs[0:n_fft//2]

plt.plot(freqs, mag, "-")
plt.grid(1);
# plt.xlim(0, 1);
```
