Implementar um sistema
$$
H(s)=\frac{s+1}{s+2s+9}
$$

```python
import numpy as np, scipy as sp, matplotlib.pyplot as plt
from scipy import signal # importa biblioteca de sinais

# definir o sistema
num = [1, 1]
den = [1, 2, 9]
sys = signal.lti(num, den)
```

```python
# resposta ao impulso do sistema
t, y = signal.impulse(sys)

plt.figure(figsize=(8,3))
plt.plot(t, y)
plt.title('Resposta ao impulso de $H(s)$')
plt.xlabel('tempo');
plt.ylabel('$y(t)$');
```

```python
# resposta ao degrau do sistema
t, y = signal.step(sys)

plt.figure(figsize=(8,3))
plt.plot(t, y)
plt.title('Resposta ao degrau de $H(s)$')
plt.xlabel('tempo');
plt.ylabel('$y(t)$');
```

```python
# resposta a um sinal de rudo na entrada

# construindo sinal x de ruido
tf = 30
dt = 0.1
t = np.arange(0, tf, dt)
N = len(t) # numero de amostras do vetor tempo
x = np.random.normal(100, 100, N)

t, y, _ = signal.lsim(sys, x, t)

plt.figure(figsize=(8,3))
plt.plot(t, x)
plt.title('Sinal de Entrada')
plt.xlabel('tempo');
plt.ylabel('$x(t)$');

plt.figure(figsize=(8,3))
plt.plot(t, y)
plt.title('Sinal de sada')
plt.xlabel('tempo');
plt.ylabel('$y(t)$');
```

```python
# resposta em frequencia do sistema
w, H = signal.freqresp(sys)
mag = np.abs(H)
pha = np.angle(H, deg=1)

plt.figure(figsize=(8,2))
plt.plot(w, mag)
plt.title('Resposta em Frequencia')
plt.xlabel('frequencia (rad/s)');
plt.ylabel('$|H(j\omega)|$');

plt.figure(figsize=(8,2))
plt.plot(w, pha)
plt.xlabel('frequencia (rad/s)');
plt.ylabel('$\measuredangle H(j\omega)$');
```

```python
# diagrama de Bode
w = np.logspace(-2,4,1000)
w, mag, pha = signal.bode(sys, w)

plt.figure(figsize=(8,2))
plt.semilogx(w, mag)
plt.title('Resposta em Frequencia')
plt.grid(1)
plt.xlabel('frequencia (rad/s)');
plt.ylabel('$|H(j\omega)|$');

plt.figure(figsize=(8,2))
plt.semilogx(w, pha)
plt.grid(1)
plt.xlabel('frequencia (rad/s)');
plt.ylabel('$\measuredangle H(j\omega)$');
```
